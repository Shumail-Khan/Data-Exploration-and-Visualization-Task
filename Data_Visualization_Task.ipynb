```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
# Seaborn provides the Iris dataset as a clean pandas DataFrame
df = sns.load_dataset('iris')

# 2. Inspect the data
print("--- Dataset Shape ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- Column Names ---")
print(df.columns.tolist())

print("\n--- First 5 Rows ---")
print(df.head())

# 3. Summary Statistics
print("\n--- Data Info ---")
df.info()

print("\n--- Descriptive Statistics ---")
print(df.describe())
```

### Visualization

```python
# Set the visual style for all plots
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 10))

# A. Scatter Plot: Petal Length vs Petal Width
# This relationship typically shows the clearest separation between species
plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=70)
plt.title('Petal Dimensions by Species')

# B. Histogram: Distribution of Sepal Length
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='sepal_length', kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')

# C. Box Plot: Identifying Outliers in Sepal Width
# Sepal Width is known to have some outliers in this dataset
plt.subplot(2, 2, 3)
sns.boxplot(data=df, hue='species', y='sepal_width', legend=False, palette='Set2')
plt.title('Sepal Width Outliers by Species')

# D. Pair Plot (Bonus): Shows all relationships at once
# Note: This creates its own figure, so we call it separately
plt.tight_layout()
plt.show()

# Run this to see the full matrix of relationships
sns.pairplot(df, hue='species', height=2.5)
plt.show()
```
